```python
quiz = [
    {
         "S no" : 1,
         "question" : "What is your name",
         "Options" : ["Ali","Aslam","Ahmed","Akber"],
        "answer" : "Ali"
         
    },
    {
        "S no" : 2,
        "question" : "What is your father name",
        "options" : ["waleed","Afzal","inayat","shoaib"],
        "answer" :" waleed"
    },
    {
        "S no" : 3,
       "question" : "In which country do you live",
        "options" : ["pakistan","USA","Africa","Uae"],
        "answer" : "pakistan"
        
    }
]

score = 0

for question in quiz :
    data = str(question["S no."]) + " : " + question["question"] + " " + question["options"][1] + " " + question["options"][2] + " " + question["options"][3]
    answer = input(data)
    if answer == question["answer"]:
        score += 1
    print("your score is:" +str(score))
```
